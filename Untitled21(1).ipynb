```python
def remplazar(fila , cadena):
    r = re.findall(cadena , fila)
    if not r:
        return fila
    for dato in r:
        fila = re.sub(dato , "" ,fila)
        return fila
```

```python
df.head(20)
```

```python
df["twiter limpiado"] = np.vectorize(remplazar)(df["tweet"],"@[\w]*")
```

```python
df.head(60)
```

```python
import re

df["twiter limpiado"] = df["twiter limpiado"].apply(lambda x: re.sub(r'[^a-zA-Z#\s]', '', x))

```

```python
df.head(40)
```

```python
def palabras(texto):
    palabras = texto.split()
    palabras_largas= [palabra for palabra in palabras if len(palabra) > 3]
    return " ".join(palabras_largas)
            
```

```python
df["twiter limpiado"] = df["twiter limpiado"].apply(palabras)
```

```python
df.head(90)
```

```python
#tokenizacion
tweet_tokenizado = df["twiter limpiado"].apply(lambda x:x.split())
tweet_tokenizado.head()
```

```python
from nltk.stem.porter import PorterStemmer
stemmer = PorterStemmer()
tweet_tokenizado = tweet_tokenizado.apply(lambda sentence: [stemmer.stem(word) for word in sentence])
tweet_tokenizado.head()
```

```python
for i in range(len(tweet_tokenizado)):
    tweet_tokenizado[i] = " ".join(tweet_tokenizado[i])
    df["twiter limpiado"] = tweet_tokenizado
    df.head()
```

```python
def removeht(tweets):
    hashtags = []
    for tweet in tweets:
        hs = re.findall(r"#(\w+)" , tweet)
        hashtags.append(hs)
    return hashtags
```

```python
ht_positive = removeht(df["twiter limpiado"][df["label"] == 0])
```

```python
ht_negative = removeht(df["twiter limpiado"][df["label"] == 1])
```

```python
ht_positive
```

```python
ht_positive = sum(ht_positive,[])
ht_negative = sum(ht_negative,[])
```

```python
frecuencia = nltk.FreqDist(ht_positive)
hp = pd.DataFrame({"Nombre" :list(frecuencia.keys()),
                   "count" : list(frecuencia.values())})
```

```python
hp.head()
```

```python
hp = hp.nlargest(columns = "count" , n = 10)
plt.figure(figsize=(20,10))
sns.barplot(data = hp , x= "Nombre" ,y="count")
plt.show()
```

```python
frecuencia = nltk.FreqDist(ht_negative)
tn = pd.DataFrame({"tweet" : list(frecuencia.keys()),
                   "repeticiones" :list(frecuencia.values())})
```

```python
tn.head()
```

```python
tn = tn.nlargest(columns = "repeticiones" , n = 10)
plt.figure(figsize=(25,12))
sns.barplot(data=tn , x="tweet" ,y="repeticiones")
plt.show()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
bowvectorizer = CountVectorizer(max_df = 0.90 , min_df = 2 , max_features = 1000 , stop_words ="english")
bow =bowvectorizer.fit_transform(df["twiter limpiado"])
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(bow , df["label"] )
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score , f1_score

lr = LogisticRegression()
lr.fit(x_train , y_train)
```

```python
ypred = lr.predict(x_test)
```

```python
accuracy_score(ypred,y_test)
```
